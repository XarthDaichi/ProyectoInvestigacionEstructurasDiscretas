```python
#código basado en https://programmerclick.com/article/21021656414/
def fuerza_bruta(texto, patron):
    n = len(texto)
    m = len(patron)
    i = 0
    while i < n-m +1 :
        j = 0
        while j < m:
            if texto[i + j] != patron[j]:
                break

            j += 1
        if j == m:
            return i
        i += 1
    return -1

txt = "abcde"
pat = "cde"
print(fuerza_bruta(txt, pat))

txt = "abcde"
pat = "123"
print(fuerza_bruta(txt, pat))

txt = "ABABDABACDABABCABAB"
pat = "ABABCABAB"
print(fuerza_bruta(txt, pat))

txt="abcdecde"
pat="cde"
print(fuerza_bruta(txt, pat))
```

```python
import matplotlib.pyplot  as plt
import numpy as np
import math as m

def fuerza_bruta_instrumentado(texto, patron):
    operaciones = 0
    # operaciones = y len() x2
    operaciones += 4
    n = len(texto)
    m = len(patron)
    # el =
    operaciones += 1
    i = 0
    
    while i < n-m +1 :
        # el < el - y el + de antes
        operaciones += 3
        # el = de abajo
        operaciones += 1
        j = 0
        
        while j < m:
            # operaciones <
            operaciones += 1
            # operaciones [] y !=
            operaciones += 3
            if texto[i + j] != patron[j]:
                break
            # operacion +=
            operaciones += 1
            j += 1
        # operaciones ==
        operaciones += 1
        if j == m:
            return operaciones
        # operacion +=
        operaciones += 1
        i += 1
    return operaciones

def test_fuerza_bruta_instrumentado(start, end, inc):
    graph = []
    graph.append([])
    graph.append([])
    graph.append([])
    for i in range(start, end, inc):
        for j in range(i, end, inc):
            a = "-" * i
            b = "+" * j + "-" * i
            graph[0].append(j+i)
            graph[1].append(fuerza_bruta_instrumentado(b,a))
    return graph

g = test_fuerza_bruta_instrumentado(10, 200, 10)
plt.plot(g[0], g[1], color="blue", label="fuerza_bruta") 
plt.legend()
plt.xlabel('Time')
plt.ylabel("N")
plt.title("Análisis de Tiempo de Fuerza Bruta")
plt.show()
```

```python
def brute_force(txt,pat):
    n = len(txt)
    m = len(pat)
    ti, pi, tj = 0, 0, 0
    newList = []
    while(ti <= n - m):
        if txt[ti] == pat[0]:
            pi = 0
            tj = ti
            while txt[tj] == pat[pi]:
                tj += 1
                pi += 1
                if pi == m:
                    newList.append(ti)
                    break
        ti += 1
    return newList

txt = "abcde"
pat = "cde"
print(brute_force(txt, pat))

txt = "abcde"
pat = "123"
print(brute_force(txt, pat))

txt = "ABABDABACDABABCABAB"
pat = "ABABCABAB"
print(brute_force(txt, pat))

txt="abcdecde"
pat="cde"
print(brute_force(txt, pat))
```
