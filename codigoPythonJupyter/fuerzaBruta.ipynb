```python
def fuerza_bruta(texto, patron):
    n = len(texto)
    m = len(patron)
    for i in range((n-m) +1):
        for j in range(m):
            if texto[i + j] != patron[j]:
                break
        if j == m-1:
            return i
    return -1

txt = "abcde"
pat = "cde"
print(fuerza_bruta(txt, pat))

txt = "abcde"
pat = "123"
print(fuerza_bruta(txt, pat))

txt = "ABABDABACDABABCABAB"
pat = "ABABCABAB"
print(fuerza_bruta(txt, pat))
```

```python
import matplotlib.pyplot  as plt
import numpy as np
import math as m

def fuerza_bruta_instrumentado(texto, patron):
    operaciones = 0
    n = len(texto)
    m = len(patron)
    for i in range((n-m) +1):
        for j in range(m):
            # operaciones [] y !=
            operaciones += 3
            if texto[i + j] != patron[j]:
                break
        if j == m-1:
            return operaciones
    return operaciones

def test_fuerza_bruta_instrumentado(start, end, inc):
    graph = []
    graph.append([])
    graph.append([])
    graph.append([])
    for i in range(start, end, inc):
        for j in range(i, end, inc):
            a = "-" * i
            b = "-"*j
            graph[0].append(i+j)
            graph[1].append(fuerza_bruta_instrumentado(a,b))
    return graph

g = test_fuerza_bruta_instrumentado(10, 200, 10)
plt.plot(g[0], g[1], color="blue", label="fuerza_bruta") 
plt.legend()
plt.xlabel('Time')
plt.ylabel("N")
plt.title("Análisis de Tiempo de merge")
plt.show()
```
